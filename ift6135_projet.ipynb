## 4.1 VISUALIZING THE BENEFITS OF CERTIFIED ROBUSTNESS
### Synthetic data

```python
train_x, train_y = main.synthetic_data(10000)
valid_x, valid_y = main.synthetic_data(4000)

batch_size = 128
loss_function = nn.CrossEntropyLoss()

train_data = torch.utils.data.TensorDataset(
        torch.from_numpy(train_x).float(), torch.from_numpy(train_y).long())
valid_data = torch.utils.data.TensorDataset(
        torch.from_numpy(valid_x).float(), torch.from_numpy(valid_y).long())
train_data_loader = torch.utils.data.DataLoader(
                train_data, batch_size=batch_size, shuffle=True, num_workers=2)
valid_data_loader = torch.utils.data.DataLoader(
                valid_data, batch_size=batch_size, shuffle=True, num_workers=2)
```

#### Train models of  WRM, ERM, FGM and IFGM with activation RELU

```python
GAMMA = 2
LR0 = 0.01 # learning rate for classify
MAX_LR0=0.08 # learning rate for maximize the perturbation, and for IFGM's step size also.
# WRM
net_WRM = main.MLP(activation='relu')
net_WRM.init_weights_glorot()
optimizer = torch.optim.Adam(net_WRM.parameters(), lr=LR0)
main.train_WRM(net_WRM,optimizer,loss_function, train_data_loader,valid_data_loader, 30 ,GAMMA,  max_lr0=MAX_LR0, min_lr0=LR0, min_lr_adjust=False)
```

```python
EPSILON=0.26646 #0.03 #0.3
# rho = 0.071, epsilon = sqrt(rho) = 0.26646 (RELU)

# ERM
net_ERM = main.MLP('relu')
net_ERM.init_weights_glorot()
optimizer = torch.optim.Adam(net_ERM.parameters(), lr=LR0)
main.train(net_ERM,optimizer,loss_function, train_data_loader,valid_data_loader,30,min_lr0=LR0,min_lr_adjust=False)
# FGM
net_FGM = main.MLP('relu')
net_FGM.init_weights_glorot()
optimizer = torch.optim.Adam(net_FGM.parameters(), lr=LR0)
main.train_FGM(net_FGM,optimizer,loss_function, train_data_loader,valid_data_loader, 30, epsilon=EPSILON, min_lr0=LR0,min_lr_adjust=False)
# IFGM
net_IFGM = main.MLP(activation='relu')
net_IFGM.init_weights_glorot()
optimizer = torch.optim.Adam(net_IFGM.parameters(), lr=LR0)
main.train_IFGM(net_IFGM,optimizer,loss_function, train_data_loader,valid_data_loader, 30, epsilon=EPSILON,min_lr0=LR0,alpha=MAX_LR0,min_lr_adjust=False)
```

#### Train models of  WRM, ERM, FGM and IFGM with activation ELU

```python
GAMMA = 2
LR0 = 0.01 # learning rate for classify
MAX_LR0=0.08 # learning rate for maximize the perturbation, and for IFGM's step size also.
EPSILON=0.2608 #0.03 #0.3
# rho = 0.068, epsilon = sqrt(rho) =0.2608 (ELU) 

# ERM
net_ERM = main.MLP('elu')
net_ERM.init_weights_glorot()
optimizer = torch.optim.Adam(net_ERM.parameters(), lr=LR0)
main.train(net_ERM,optimizer,loss_function, train_data_loader,valid_data_loader,30,min_lr0=LR0,min_lr_adjust=False)
# FGM
net_FGM = main.MLP('elu')
net_FGM.init_weights_glorot()
optimizer = torch.optim.Adam(net_FGM.parameters(), lr=LR0)
main.train_FGM(net_FGM,optimizer,loss_function, train_data_loader,valid_data_loader, 30, epsilon=EPSILON, min_lr0=LR0,min_lr_adjust=False)
# IFGM
net_IFGM = main.MLP(activation='elu')
net_IFGM.init_weights_glorot()
optimizer = torch.optim.Adam(net_IFGM.parameters(), lr=LR0)
main.train_IFGM(net_IFGM,optimizer,loss_function, train_data_loader,valid_data_loader, 30, epsilon=EPSILON,min_lr0=LR0,alpha=MAX_LR0,min_lr_adjust=False)

# WRM
net_WRM = main.MLP(activation='elu')
net_WRM.init_weights_glorot()
optimizer = torch.optim.Adam(net_WRM.parameters(), lr=LR0)
main.train_WRM(net_WRM,optimizer,loss_function, train_data_loader,valid_data_loader, 30 ,GAMMA,  max_lr0=MAX_LR0, min_lr0=LR0, min_lr_adjust=False, savepath='syn_WRM')
```

### Certiﬁcate against the out-of-sample (MNIST test) worst-case performance $sup_{P:W_c(P,P_0)≤\rho} E_P[l(\theta;Z)]$ for WRM with ELU’s

```python
model = mnist.Mnist_Estimateur()
#mnist_certificate_train_data
#mnist_certificate2
#mnist_certificate_test_data
net_WRM, list_certi = main.loadCheckpoint(model,'mnist_certificate_test_data')
certificate=list_certi['list_certi']
list_rho=list_certi['list_rho']
list_worst=list_certi['list_worst']
plt.plot(list_rho,list_worst, c='red', label=r"Test worst-case: $\sup_{P:W_c(P,\hat{P}_{test}) \leq \rho } E_P [l(\theta_{WRM};Z)]$")
plt.plot(np.array(range(0,420,30))/100.0,certificate,c='blue', label=r"Certificate: $E_{\hat{P}_n}[\phi_{\gamma}(\theta_{WRM};Z)]+\gamma \rho$")
# plt.plot(np.array(range(0,400,50))/100.0,certificate,c='blue', label=r"Certificate: $E_{\hat{P}_n}[\phi_{\gamma}(\theta_{WRM};Z)]+\gamma \rho$")
last_rho_train=0.312
plt.plot([last_rho_train,last_rho_train],[0.0,1.6], linestyle="--", c="black") 
plt.xlabel(r"$\rho$")
plt.xlim([0.0,3.6])
plt.ylim([0.0,2.0])
plt.xticks([0,0.5,1,1.5,2,2.5,3,3.5])
plt.legend(loc="lower right")
plt.savefig('fig2-mnist-t15-ep30.pdf')
```

### PGM attacks on the MNIST dataset.
#### (a) Test error vs. $\epsilon_{adv}$ for $||·||_2$ attack

```python
model = mnist.Mnist_Estimateur()
_,list_p2= main.loadCheckpoint(model,'mnist_attack_pgm_p2_list_errs')

epsilons = np.array(range(0,27,2))/100.0
labels =['ERM','FGM','IFGM','WRM']

plt.plot(epsilons,list_p2[0],linestyle='-', marker ='.',label=labels[0])
plt.plot(epsilons,list_p2[1],linestyle='-', marker ='*',label=labels[1])
plt.plot(epsilons,list_p2[2],linestyle='-', marker ='>',label=labels[2])
plt.plot(epsilons,list_p2[3],linestyle='-', marker ='<',label=labels[3])
plt.plot([0.5586/9.21,0.5586/9.21],[0.0,0.6], linestyle="--", c="black") 
plt.yscale('log')
plt.xlabel(r"$\epsilon_{adv}/C_{2}$")
plt.ylabel(r"$Error$")
# plt.xticks([0.05])
# plt.title(r"$\hat\rho_{test}$ vs. $1/\gamma_{adv}$")
plt.legend()
plt.savefig('figure3_attack_pgm_p2.pdf')
plt.show()
# plt.close()
```
